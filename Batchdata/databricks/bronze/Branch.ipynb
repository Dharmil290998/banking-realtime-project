```python
from pyspark.sql.functions import *
```

```python
df_branch = (
    spark.readStream
    .format("cloudFiles")
    .option("cloudFiles.format", "csv")
    .option("header", "true")

    .option("cloudFiles.useNotifications", "true")
    .option("cloudFiles.subscriptionId", "4967f902-d637-449d-81c4-901cd225af2f")
    .option("cloudFiles.tenantId", tenant_id)
    .option("cloudFiles.clientId", client_id)
    .option("cloudFiles.clientSecret", secret_value)
    .option("cloudFiles.includeExistingFiles", "true")
    .option("cloudFiles.resourceGroup", "Realtimedata")
    .option("cloudFiles.queueName", "autoloader-branch-queue")
    .option("cloudFiles.schemaEvolutionMode", "addNewColumns")
    .option("cloudFiles.schemaLocation", "/Volumes/bankingdata/bronze/schemalocation/branch/")
    .option("badRecordsPath", "/Volumes/bankingdata/bronze/badrecords/branch/")
    .option("pathGlobFilter", "*.csv")

    .load("abfss://project2@project1demo.dfs.core.windows.net/branch/")
)

df_branch = df_branch.withColumn("ingestion_time", current_timestamp()) \
                     .withColumn("source_file", input_file_name())

df_branch.writeStream \
    .format("delta") \
    .option("checkpointLocation", "/Volumes/bankingdata/bronze/checkpoints/branch/") \
    .outputMode("append") \
    .trigger(availableNow=True) \
    .toTable("bankingdata.bronze.branch")
```
