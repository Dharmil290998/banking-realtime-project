```python
df = spark.read.table("bankingdata.silver.transaction")
```

```python
df = df.select(
    "transaction_id",
    "account_id",
    "customer_id",
    "branch_id",
    "card_id",
    "product_id",
    "transaction_type",
    "amount",
    "merchant_category",
    "location",
    "event_time",
    "is_high_value"
)
```

```python
df.write \
    .format("delta") \
    .mode("append") \
    .saveAsTable("bankingdata.gold.fact_Transaction")
```
